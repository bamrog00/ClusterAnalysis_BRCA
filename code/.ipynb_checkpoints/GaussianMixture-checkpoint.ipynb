```python
def gmmHRD(score_type):
    '''
    Function descr
    Intput: score_type (dataframe)
    Output: cutoff (Int), likelihood (float)
    '''
    minimums = []
    score_type = score_type[['case_id','HRD_sum']]
    score_type.set_index('case_id', inplace = True)
    
    gmm = GaussianMixture(n_components=2).fit(score_type)
    labels = gmm.predict(score_type)
    likelihood = gmm.score(score_type)
    aic = gmm.aic(score_type)
    bic = gmm.bic(score_type)
    score_type['labels'] = labels
    minimums.append(min(score_type[score_type['labels'] == 0]['HRD_sum']))
    minimums.append(min(score_type[score_type['labels'] == 1]['HRD_sum']))
    cutoff = max(minimums)
    
    return cutoff, likelihood, aic, bic

def gmmHRD_cf(score_type,n_folds):
    
    score_type = score_type[['case_id','HRD_sum']]
    score_type.set_index('case_id', inplace = True)
    
    # create a KFold object with the specified number of folds
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

    
    likelihoods = []
    aics = []
    bics = []
    cutoffs = []

    
    for fold, (train_idx, test_idx) in enumerate(kf.split(score_type)):
        minimums = []
        
        train_data = score_type.iloc[train_idx]
        test_data = score_type.iloc[test_idx]

        
        gmm = GaussianMixture(n_components=2)

        gmm.fit(train_data)

        likelihood = gmm.score(test_data)
        aic = gmm.aic(test_data)
        bic = gmm.bic(test_data)
        
        
        likelihoods.append(likelihood)
        aics.append(aic)
        bics.append(bic)
       
    
        labels = gmm.predict(test_data)

        # get the parameters of each component for this fold
        means = gmm.means_
        covariances = gmm.covariances_
        weights = gmm.weights_

        # create a range of values for the x-axis
        x = np.linspace(test_data['HRD_sum'].min(), test_data['HRD_sum'].max(), 1000)

        test_data['label'] = labels
        try: 
            minimums.append(min(test_data[test_data['label'] == 0]['HRD_sum']))
            minimums.append(min(test_data[test_data['label'] == 1]['HRD_sum']))
        except:
            print('Encountered empty sequence')
            continue
        cutoffs.append(max(minimums))
    
    return np.mean(cutoffs), np.mean(likelihoods), np.std(likelihoods), np.mean(aics), np.std(aics), np.mean(bics), np.std(bics), len(cutoffs)


def gmm_commponents(score_type):
   
    score_type = score_type[['case_id','HRD_sum']]
    score_type.set_index('case_id', inplace = True)
    
    
    results = dict()
    n_components = [2,3]
    
    for n in n_components:
        minimums = []
        gmm = GaussianMixture(n_components=n).fit(score_type)
        labels = gmm.predict(score_type)
        likelihood = gmm.score(score_type)
        aic = gmm.aic(score_type)
        bic = gmm.bic(score_type)
        score_type['labels'] = labels
        if n == 2:
            minimums.append(min(score_type[score_type['labels'] == 0]['HRD_sum']))
            minimums.append(min(score_type[score_type['labels'] == 1]['HRD_sum']))
            results['n2_cutoff'] = max(minimums)
            results['n2_likelihood'] = likelihood
            results['n2_aic'] = aic
            results['n2_bic'] = bic
        else:
            label0 = score_type[score_type['labels'] == 0]['HRD_sum']
            label1 = score_type[score_type['labels'] == 1]['HRD_sum']
            label2 = score_type[score_type['labels'] == 2]['HRD_sum']
            results['n3_likelihood'] = likelihood
            results['n3_aic'] = aic
            results['n3_bic'] = bic
            if label0.shape[0] == 0:
                results['Labels'] = 'Yes_lab0'
                minimums.append(min(score_type[score_type['labels'] == 1]['HRD_sum']))
                minimums.append(min(score_type[score_type['labels'] == 2]['HRD_sum']))
                minimums.sort()
                results['n3_cutoff_1'] = minimums[-2]
                results['n3_cutoff_2'] = minimums[-1]
            elif label1.shape[0] == 0:
                results['Labels'] = 'Yes_lab1'
                minimums.append(min(score_type[score_type['labels'] == 0]['HRD_sum']))
                minimums.append(min(score_type[score_type['labels'] == 2]['HRD_sum']))
                minimums.sort()
                results['n3_cutoff_1'] = minimums[-2]
                results['n3_cutoff_2'] = minimums[-1]
            elif label2.shape[0] == 0:
                results['Labels'] = 'Yes_lab2'
                minimums.append(min(score_type[score_type['labels'] == 0]['HRD_sum']))
                minimums.append(min(score_type[score_type['labels'] == 1]['HRD_sum']))
                minimums.sort()
                results['n3_cutoff_1'] = minimums[-2]
                results['n3_cutoff_2'] = minimums[-1]
            else:
                minimums.append(min(score_type[score_type['labels'] == 0]['HRD_sum']))
                minimums.append(min(score_type[score_type['labels'] == 1]['HRD_sum']))
                minimums.append(min(score_type[score_type['labels'] == 2]['HRD_sum']))
                minimums.sort()
                results['n3_cutoff_1'] = minimums[-2]
                results['n3_cutoff_2'] = minimums[-1]
                results['Labels'] = 'No'


    return results
```

```python
df = pd.DataFrame(columns=['Project ID','n_cases', 'cutoff', 'likelihood', 'aic', 'bic', 'n_high','n_low'])
types = np.unique(HRD_scores_pan_cancer_primary['Project ID'])

for cancer_type in types:
    score_type = HRD_scores_pan_cancer_primary[HRD_scores_pan_cancer_primary['Project ID'] == cancer_type]
    cutoff, likelihood, aic, bic = gmmHRD(score_type.copy())
    #df.append({'Column1': i, 'Column2': i*2, 'Column3': i*3}, ignore_index=True)
    high = score_type[score_type['HRD_sum']>=cutoff]
    low = score_type[score_type['HRD_sum']<cutoff]
    df.loc[len(df)] = [cancer_type, score_type.shape[0], cutoff, likelihood, aic, bic, high.shape[0], low.shape[0]]
    
```

```python
df.to_csv('../data/gmm_cutoffs.csv', sep=',', header = True)
```

```python
df = pd.DataFrame(columns=['Project ID','n_cases','n_fold', 'cutoff_mean', 'likelihood_mean', 'likelihood_std', 'aic_mean', 'aic_std', 'bic_mean', 'bic_std'])
types = np.unique(HRD_scores_pan_cancer_primary['Project ID'])

for cancer_type in types:
    n_folds = 10
    score_type = HRD_scores_pan_cancer_primary[HRD_scores_pan_cancer_primary['Project ID'] == cancer_type]
    if score_type.shape[0] < n_folds:
        if score_type.shape[0] < (n_folds/2):
            n_folds = int(score_type.shape[0])
        else:
            n_folds = int(n_folds/2)
    cutoff, likelihood_mean, likelihood_std, aic_mean, aic_std, bic_mean, bic_std, n_folds = gmmHRD_cf(score_type,n_folds=n_folds)
    
    df.loc[len(df)] = [cancer_type, score_type.shape[0], n_folds, cutoff, likelihood_mean, likelihood_std, aic_mean, aic_std, bic_mean, bic_std]
```

```python
df.to_csv('../data/gmm_cutoffs_cf.csv', sep=',', header = True)
```

```python
df = pd.DataFrame(columns=['Project ID','n_cases', 'n2_cutoff', 'n2_likelihood', 'n2_aic', 'n2_bic','n3_cutoff_1','n3_cutoff_2','n3_likelihood','n3_aic','n3_bic','empty_set'])
types = np.unique(HRD_scores_pan_cancer_primary['Project ID'])

for cancer_type in types:
    score_type = HRD_scores_pan_cancer_primary[HRD_scores_pan_cancer_primary['Project ID'] == cancer_type]
    results = gmm_commponents(score_type.copy())
    df.loc[len(df)] = [cancer_type, score_type.shape[0], results['n2_cutoff'], results['n2_likelihood'], results['n2_aic'], results['n2_bic'],results['n3_cutoff_1'],results['n3_cutoff_2'],results['n3_likelihood'], results['n3_aic'], results['n3_bic'],results['Labels']]
```
