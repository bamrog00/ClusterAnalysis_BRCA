```python
feature_counts_brca012715_col.pop(0)
feature_counts_brca06015_col.pop(0)
raw_data_firebrowse_col.pop(0)
firebrowse_RSME_col.pop(0)
```

```python
def shortenID(id_list):
    shorten_ids = list()
    for id_ in id_list:
        split = id_.split('-')[0:4]
        shorten_ids.append('-'.join(split))
    return shorten_ids
```

```python
shorten_feature_counts_brca012715_col = shortenID(feature_counts_brca012715_col)
shorten_feature_counts_brca06015_col = shortenID(feature_counts_brca06015_col)
shorten_normalized_results_brca_col = shortenID(normalized_results_brca_col)
shorten_raw_data_firebrowse_col = shortenID(raw_data_firebrowse_col)
shorten_firebrowse_RSME_col = shortenID(firebrowse_RSME_col)
```

```python
sample_id = list()
for sampleid in sample_id_double:
    ids = sampleid.split(', ')
    id_1 = ids[0]
    id_2 = ids[1]
    if id_1 == id_2:
        sample_id.append(id_1)
    else:
        sample_id.append(id_1)
        sample_id.append(id_2)
```

```python
def checkMatches(list1,list2):
    set1 = set(list1)
    set2 = set(list2)
    
    matching = list(set1.intersection(set2))
    not_matching = list(set1.symmetric_difference(set2))

    return matching, not_matching
```

```python
matching, _ = checkMatches(sample_id,shorten_feature_counts_brca012715_col)
print('HRD-files (' + str(len(sample_id)) + ' IDs) compared to fc012715 (' + str(len(np.unique(np.array(shorten_feature_counts_brca012715_col)))) + ' unique ID/cases): ' + str(len(matching)) + '/' + str(len(sample_id_double)))

matching, _ = checkMatches(sample_id,shorten_feature_counts_brca06015_col)
print('HRD-files (' + str(len(sample_id)) + ' IDs) compared to fc06015 (' + str(len(np.unique(np.array(shorten_feature_counts_brca06015_col)))) + ' unique ID/cases): ' + str(len(matching)) + '/' + str(len(sample_id_double)))

matching, _ = checkMatches(sample_id,shorten_normalized_results_brca_col)
print('HRD-files (' + str(len(sample_id)) + ' IDs) compared to normalized_results (' + str(len(np.unique(np.array(shorten_normalized_results_brca_col)))) + ' unique ID/cases): ' + str(len(matching)) + '/' + str(len(sample_id_double)))

matching, _ = checkMatches(sample_id,shorten_raw_data_firebrowse_col)
print('HRD-files (' + str(len(sample_id)) + ' IDs) compared to firebrowse (' + str(len(np.unique(np.array(shorten_raw_data_firebrowse_col)))) + ' unique ID/cases): ' + str(len(matching)) + '/' + str(len(sample_id_double)))

matching, _ = checkMatches(sample_id,shorten_firebrowse_RSME_col)
print('HRD-files (' + str(len(sample_id)) + ' IDs) compared to firebrowseRSME (' + str(len(np.unique(np.array(shorten_firebrowse_RSME_col)))) + ' unique ID/cases): ' + str(len(matching)) + '/' + str(len(sample_id_double)))
```
